# Shared interests vs shared racial background

```python
#Ne garder que les colonnes sur l'origine, la dcision et les intrts communs
data_light = data.loc[:, ['samerace', 'dec', 'shar']]
data_light.describe()
```

```python
#Supprimer les lignes ne possdant pas de valeurs pour "shar"
data_light = data_light.dropna(subset=["shar"])
data_light.describe()
```

Note:
On observe que la dcision est 43.3% un accord et donc 56.7% un refus. Une baseline de prdiction serait donc de 56.7%.

```python
#Sparer les variables de la target
X = data_light[['samerace', 'shar']].copy()
y = data_light['dec']

#Standardiser les donnes
scaler = StandardScaler()
X["shar"] = scaler.fit_transform(X[["shar"]])

#Ajouter l'intercept
X = sm.add_constant(X)

#Fit le modle
model = sm.Logit(y, X).fit()

#Afficher les rsultats
print(model.summary())

#Calculer l'accuracy
y_pred = (model.predict(X) > 0.5).astype(int)
print(f"Accuracy: {(y_pred == y).mean():.1%} | Baseline: {(1-y.mean()):.1%}")
```

Note:
On a standardis la variable 'shar', nous pouvons donc maintenant la comparer avec la variable samerace. De plus le modle a une accuracy de 67.9%, ce qui est bien meilleur que la baseline de 56.7%, ce qui montre que les features permettent rellement de prdire la dcision.

Les coefficients:
- Le coefficient de samerace est de 0.0165, donc avoir les mmes origines n'impacte quasiment pas l'accord d'un second rendez-vous.
- Le coefficient de shar est de 0.9627, c'est environ 58 fois plus important que samerace.

Les p-values:
- La p-value de samerace est de 0.754. Donc, si samerace n'avait aucun effet rel, il y aurait 75.4% de chances d'observer un coefficient au moins aussi grand que 0.0165, juste par hasard. L'effet de samerace n'est pas statistiquement significatif.
- La p-value de shar est de 0. Il y a moins de 0.001% de chances d'obtenir un coefficient de 0.9627 si shar n'avait aucun effet. L'effet de shar est trs significatif.

Les intervalles de confiance:
- L'intervalle de confiance de samerace est [-0.087, 0.120]. On ne peut pas dire si l'effet de samerace est positif, ngatif ou mme nul. En tout cas, avec 95% de certitude, il a un impact trs modr.
- L'intervalle de confiance de shar est de [0.903, 1.022]. Mme dans le pire des cas, le coefficient est trs lev. Donc on est sr  95% que l'effet d'une note leve en intrts communs a un effet positif et fort sur l'acceptation d'un second rendez-vous.

## Conclusion
Les origines ont un impact si petit sur la prise de dcision qu'on ne peut mme pas affirmer qu'il existe. Donc malgr l'incertitude sur leur impact (positif/ngatif/nul), les origines ont bien moins d'effet que les intrts communs qui, eux, ont un impact fort sur l'accord d'un second rendez-vous.

# Diffrences des dcisions entre le premier et le dernier rendez-vous

```python
#Filtrer en ne gardant que les colonnes nous intressant et supprimer les lignes contenant des valeurs manquantes
clean_data = data[["iid", "pid", "order", "dec_o"]].dropna()

print("Nombre de dates aprs nettoyage: ", len(clean_data))
print(clean_data['order'].describe())

print(f"\nNombre de dates  l'ordre 1 (premiers): {len(clean_data[clean_data['order']==1])}")
print(f"Taux d'acceptation quand on est le premier date: {clean_data[clean_data['order']==1]['dec_o'].mean():.1%}")
```

```python
#Trouver le dernier ordre de chaque individu
last_orders = clean_data.groupby('iid')['order'].max().reset_index()
last_orders.columns = ['iid', 'last_order']

#Fusionner pour identifier les derniers dates
clean_data_with_last = clean_data.merge(last_orders, on='iid')
clean_data_with_last['is_last'] = (clean_data_with_last['order'] == clean_data_with_last['last_order'])

#Filtrer sur les derniers dates
last_dates = clean_data_with_last[clean_data_with_last['is_last']]

print("Nombre de dates  l'ordre 'dernier': ", len(last_dates))
print(f"Taux d'acceptation quand on est le dernier date: {last_dates['dec_o'].mean():.1%}")
```

Note:
On observe une diffrence entre le taux d'acceptation lors d'un premier et d'un dernier rendez-vous (3.9 points).

```python
def get_statistics(last=0):
    
    #Extraire les premiers rendez-vous
    first_dates = clean_data_with_last[clean_data_with_last['order'] == 1]

    #Extraire les rendez-vous d'un certain ordre
    if last == 0: #On mesure en fonction de tous les derniers rendez-vous
        last = last_dates
    else: #On donne un nombre prcis pour l'ordre'
        last = clean_data_with_last[clean_data_with_last['order'] == last]

    #Nombre de succs
    successes = [first_dates['dec_o'].sum(), last['dec_o'].sum()]

    #Nombre total d'observations
    n_obs = [len(first_dates), len(last)]

    #Test de proportions
    z_stat, p_value = proportions_ztest(successes, n_obs)

    print(f"Premier: {first_dates['dec_o'].mean():.1%} ({successes[0]}/{n_obs[0]})")
    print(f"Autre: {last['dec_o'].mean():.1%} ({successes[1]}/{n_obs[1]})")
    print(f"\nDiffrence: {(first_dates['dec_o'].mean() - last['dec_o'].mean())*100:.1f} points")
    print(f"P-value: {p_value:.4f}")
```

```python
print("Test confondant tous les derniers rendez-vous:\n")
get_statistics()
```

```python
print("Test se concentrant sur les rendez-vous d'ordre 10:\n")
get_statistics(10)
```

## Conclusion
La diffrence observe entre un premier date et un dernier (3.9 points) n'est pas statistiquement significatives (p-value largement suprieure  0.05). On ne peut donc exclure que cette diffrence soit simplement due au hasard.
Il ne semble donc ni meilleur ni pire d'tre le premier ou le dernier rendez-vous de la soire pour obtenir un second date.

On observera qu'il existe une diffrence statistique entre tre le premier rendez-vous de quelqu'un et tre son dixime (indpendamment du fait qu'il s'agisse ou non de son dernier rendez-vous). On peut donc affirmer qu'il est prfrable d'tre le premier rendez-vous de quelqu'un plutt que son dixime pour obtenir un second rendez-vous.
